```python
scores = pd.read_pickle('data/clf_scores.pickle')
scores = scores.sort(columns=['acc'])
# break scores in to transformers and classifiers
def claim(df, st):
    return df[[s.lower().startswith(st.lower()) for  s in df['name']]]

# this has seriously gotta be the worst possible way
scores['pca'] = [s.startswith('PCA') for  s in scores['name']]
pca = scores[scores['pca']]
scores['sp'] = [s.startswith('SP') for  s in scores['name']]
sp = scores[scores['sp']]
scores['lsvc'] = [s.startswith('LSVC') for  s in scores['name']]
lsvc = scores[scores['lsvc']]
scores['rbm'] = [s.startswith('RBM') for  s in scores['name']]
rbm = scores[scores['rbm']]
scores['none'] = (scores['pca']==False) & (scores['sp']==False) & (scores['lsvc']==False) & (scores['rbm']==False)
none = scores[scores['none']]
scores['rf'] = [s.count('RF') > 0 for  s in scores['name']]
rf = scores[scores['rf']]
scores['svc-l'] = [s.count('SVC L') > 0 for  s in scores['name']]
svcl = scores[scores['svc-l']]
scores['svc-rbf'] = [s.count('SVC RBF') > 0 for  s in scores['name']]
svcrbf = scores[scores['svc-rbf']]
scores['logit'] = [s.count('LOGIT') > 0 for  s in scores['name']]
logit = scores[scores['logit']]
scores['nb'] = [s.count('NB') > 0 for  s in scores['name']]
nb = scores[scores['nb']]
```
