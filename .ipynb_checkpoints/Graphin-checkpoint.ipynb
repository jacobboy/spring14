```python
sp = np.load('data/SP.25')
Xt_train = sp.transform(X_train)
Xt_test = sp.transform(X_test)
pca = PCA(n_components=6)
pca.fit(Xt_train, y_train)
X_r = pca.transform(Xt_test)
fig, axes = matrix_scatter_pcs(True= X_r[y_test == True], False=X_r[y_test==False], tks=False)
plt.legend()
plt.show()
```

```python
# Try to select, then RBM, then forest
clf = Pipeline([('sp', SelectPercentile(f_classif, percentile=25)),
                ('rbm', BernoulliRBM(random_state=0, n_iter = 100, n_components=100, learning_rate=1)),
                ('abs', RandomForestClassifier(n_estimators=200))])
clf.fit(X_train, y_train)
y_true, y_pred = y_test, clf.predict(X_test)
f1 = f1_score(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
print("f1-score:   %0.3f" % f1)
print("accuracy:   %0.3f" % acc)
print("precision:   %0.3f" % prec)
print()
print("Detailed classification report:")
print()
print("The model is trained on the full development set.")
print("The scores are computed on the full evaluation set.")
print()
cr = classification_report(y_true, y_pred)
print(cr)
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
rf_grd = {
    'max_features':['sqrt','log2'],
    'n_estimators':[200,500]
    }
clf = GridSearchCV(RandomForestClassifier(), rf_grd, scoring='precision')
clf.fit(X_train, y_train)
y_true, y_pred = y_test, clf.predict(X_test)
f1 = f1_score(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
print("f1-score:   %0.3f" % f1)
print("accuracy:   %0.3f" % acc)
print("precision:   %0.3f" % prec)
print()
print("Detailed classification report:")
print()
print("The model is trained on the full development set.")
print("The scores are computed on the full evaluation set.")
print()
cr = classification_report(y_true, y_pred)
print(cr)
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
rf_grd = {
    'rf__max_features':['sqrt','log2'],
    'rf__n_estimators':[200]
    }
clf = Pipeline([('rbm', BernoulliRBM(random_state=0, n_iter = 100, n_components=100, learning_rate=1)),
                ('rf', RandomForestClassifier())])
clf = GridSearchCV(clf, rf_grd, scoring='precision')
clf.fit(X_train, y_train)
print clf.best_estimator_
y_true, y_pred = y_test, clf.predict(X_test)
f1 = f1_score(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
print("f1-score:   %0.3f" % f1)
print("accuracy:   %0.3f" % acc)
print("precision:   %0.3f" % prec)
print()
print("Detailed classification report:")
print()
cr = classification_report(y_true, y_pred)
print(cr)
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
rbm_rf = clf
forest = clf.best_estimator_.named_steps.get('rf')
importances = forest.feature_importances_
std = np.std([tree.feature_importances_ for tree in forest.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1][:20]
# Print the feature ranking
print("Feature ranking:")

for f in range(10):
    print("%d. %d %s (%f)" % (f + 1, indices[f], fn[indices[f]], importances[indices[f]]))

# Plot the feature importances of the forest
import pylab as pl
pl.figure()
pl.title("Feature importances")
pl.bar(range(20), importances[indices],
       color="r", yerr=std[indices], align="center")
pl.xticks(range(20), fn[indices])
pl.xlim([-1, 20])
pl.show()
```
